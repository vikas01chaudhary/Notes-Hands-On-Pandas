# Pivot Table

1. It is a more refined version of the group by function.

2. Used to summarize the data 

```python
import pandas as pd
import numpy as np
```

```python
# Reading the files

orders = pd.read_excel(r"E:\Data Science\data set\global_superstore_2016.xlsx", sheet_name="Orders")
returns = pd.read_excel(r"E:\Data Science\data set\global_superstore_2016.xlsx", sheet_name="Returns")
people = pd.read_excel(r"E:\Data Science\data set\global_superstore_2016.xlsx", sheet_name="People")
```

```python
df1 = orders.loc[:10,["Region","Sales","Profit"]]
df1
```

```python
# Example 1 :

# I want to know the sum of profit of the all region


df1.pivot_table(values ="Profit", index ="Region", aggfunc = "sum")
```

```python
# Example 2 :

# I want to know the mean of profit of the all region, cateory wise in orders Table


orders.pivot_table(values ="Profit", index ="Region", columns="Category", aggfunc = "sum")
```

```python
orders.columns
```

```python
# Example 3 :

# I want to know the sum profit & quantity of the all region & state, cateory wise in orders Table


orders.pivot_table(values =["Profit","Quantity"], index =["Region", "State"], columns="Category", aggfunc="sum")
```
