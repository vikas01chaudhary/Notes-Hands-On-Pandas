# Group By

```python
# I want to find total profit across sub-category


df1.groupby("Sub-Category") ["Profit"].sum()
```

```python
# I want to find the total profit across the region 


df.groupby("Region") ["Profit"].sum()
```

```python
# I want to find the total profit & sales across the region 


df.groupby("Region") [["Profit", "Sales"]].sum()
```

```python
# I want to find the total profit & sales across the region & Category


df.groupby(["Region","Category"]) [["Profit", "Sales"]].sum()
```

```python
# I want to find the mean profit & sales across the region & Category


df.groupby(["Region","Category"]) [["Profit", "Sales"]].mean()
```
