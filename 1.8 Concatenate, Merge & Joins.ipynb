# Merge   ---> Uses the column names to merge DF

```python
# In Pandas we use Merge to do different types of joins
```

```python
orders.columns
```

```python
# Taking only few columns from orders table/DF

df1 = orders.loc[:10,["Region","Sales","Profit"]]
df1
```

```python
df2 = people.copy()

df2
```

## Inner Join using Merge

```python
pd.merge(df1,df2, on = "Region", how ="inner")   


# on = "give common column name"

# how = "by default its inner join only"
```

## Left Join using Merge

```python
pd.merge(df1,df2, on = "Region", how ="left")   
```

## Right Join using Merge

```python
pd.merge(df1,df2, on = "Region", how ="right")   
```

## Full/Outer Join using Merge

```python
pd.merge(df1,df2, on = "Region", how ="outer")   
```

# If we have don't have same column name

```python
# Creating the copy of DF and modifying the column names of the DF

df3 = df2.copy()

df3.columns = ["Manger","Location"]

df3
```

```python
df1
```

```python
# here, in the DF1 we have column name - "Region" & the in the DF3 we have column name - "Location"

# We have to give "on" condition on the different column names & the values of the both columns are same
```

```python
pd.merge(df1,df3, left_on = "Region",right_on = "Location", how  = "inner")
```

# Joins ---> Uses the index to merge DF 


`Conditions for Joins:`

1. Index of both DF must be same, here not neccesary that the column name should be same 
2. We have to specify the right & left suffixes in the code  

```python
df1
```

```python
# Creating the DF and accessing only 1st 5 rows from the people sheet/DF

# Changing the column names of df2 from 'Person' to 'Regional Manger' and keeping 2nd column name as same 'Region'



df2 = people.copy() [:5]


df2.columns = [" Regional Manger","Region"]

df2
```

```python
# Joins the 2 DF on the index level

df1.join(df2, lsuffix = "_left", rsuffix = "_right")
```
