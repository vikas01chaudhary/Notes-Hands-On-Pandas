# Sorting 


Sorting can be done in two ways:

1. with the help of index

2. with the help of column names

### Way 1

```python
# Sorting the index in the ascending order

df1.sort_index()


# by default it sort in the ascending order
```

```python
# Sorting the index in the decending order


df1.sort_index(ascending = False)
```

### Way 2

```python
# Sorting the data using vaules

# Sort the vaules in the ascending order,
# when we sort by using values then we have to specify the column name that we have to sort


df1.sort_values(by = 'Profit')
```

```python
# Sort the vaules in the ascending order,
# when we sort by using values then we have to specify the column name that we have to sort


df1.sort_values(by = "Profit", ascending = False)
```

```python
# Sorting the Multiple values in the ascending order


df1.sort_values(["Profit","Sub-Category"])


# 2nd column always have dependency on the 1st column

# for eg: if we have profit - 14,14,15,16,14,18

#              sub-category - A,B,C,D,A,E

#   Profit      Sub-Category

#    14            A
#    14            A
#    14            D
#    15            B
#    16            C
#    18            E


```

```python
# Let's have more clearty

df1.sort_values(["Quantity","Sub-Category"])
```

```python
# So, above sort the quantity data in ascending order and we have 2 two times and with in 2--> it has sort the sub-category 
# in the alphabetic order.
```
