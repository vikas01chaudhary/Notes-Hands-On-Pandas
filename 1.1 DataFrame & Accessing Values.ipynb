# Accessing Values from DataFrame


### There is 3 ways to access values in the Pandas: 

<br>

##### 1. loc   (Location) :

-->  Helps to access values based on the names of index,columns

-->  End Position is always consider

<br>

##### 2. iloc   (Index Location) :


Same as we used to access in Numpy :
<br>

-->  Here we access the vaules based on the position

-->  Syntax for accessing 2D array:


a[ : , : ]

a[rows,columns]

a[ SP:EP, SP:EP]

* SP -> Start Position
* EP -> End Position

-->  In Indexing End Position is always Excluded


```python

df1 = pd.DataFrame({"Eng":eng,"Maths":maths,"Physics":phy})[:5]

df1
```

```python

df = pd.DataFrame({"Eng":eng,"Maths":maths,"Physics":phy},index = names)[:5]

df
```

## 1. loc   (Location)

```python
df1.loc[:,:]
```

```python
df1.loc[1:4,'Eng':'Physics']
```

```python
# In loc, we cant access the values using the position/index, we have to speficify the names of the row/column

# Here we access the values by using names of row from 1 to 4 and column names from eng to physics

# And here, the end position is considered, so only in Location Function --> end position is considered. 
```

```python
df
```

```python
# access only column maths and physics and rows from Chad to Eva


df.loc['Chad':'Eva','Maths':'Physics']
```

## Non-Sequential Accessing

```python
df
```

```python
# same as Numpy

# I want to access only Chad and Greg, eng and physics

df.loc[['Chad','Greg'],['Eng','Physics']]
```

```python
# Here we access the values by using names of rows (Chad,Greg) and column names only eng & physics
```

```python

df.loc[['Norene','Jennifer','Greg'],['Eng','Physics']]
```

## 2. iLOC (Index-Location)

```python
 df
```

```python
df.iloc[:2,1:]


# here, we access values using index position, do not use names of column/row to access the values
# Doesn't include/considered last/end position
```

```python
df.iloc[2:,:2]
```
