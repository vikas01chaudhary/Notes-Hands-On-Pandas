## To read/import the file as DataFrame



### Method 1 :

```python

# When the file is in the same/current folder/path, then we have to just give the name of the excel file




df = pd.read_csv("sample super store.csv")

df.head(2)



# head(2) gives 1st 2 rows
```

```python
# Sometimes, file has many special characters bcz of that when we use df.head(2), it will give error

# so to handle this type of error we will do


df = pd.read_csv("sample super store.csv", encoding= "latin")

df.head(2)




# Used only when we get 'Unicodedecode Error'

# encoding just handle the special characters, this is not used every time when we read the file
```
